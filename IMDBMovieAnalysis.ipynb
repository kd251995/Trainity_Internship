# Task 1 : Find the movies with the highest profit?

First we will convert values of column 'gross' and 'budget' into smaller values as we can see that these column has bigger values.

```python
df['gross'] = df['gross'].floordiv(1000000)
df['budget'] = df['budget'].floordiv(1000000)
```

```python
'''Create a new column called profit and Sorting the column using the profit '''

df['Profit'] = df['gross'] - df['budget']
profit_sorted_df = df.sort_values(by='Profit', ascending = False)
```

```python
'''Seeing Top 10 movies which recieved highest Profit'''

profit_sorted_df.head(10)['movie_title']
```

```python
'''Seeing Top 10 director which recieved highest Profit'''

profit_sorted_df.head(10)['director_name']
```

```python
'''First we will check the data having negative profit'''

negtive_df = df[df['Profit']<0]
negtive_df
```

```python
'''Data Visualization of Budget having negative impact on Profit'''

sns.jointplot(x=negtive_df['budget'],y=negtive_df['Profit'])
```

```python
'''Top 5 Movies having negative profit'''

negtive_df.sort_values(by='Profit').head()['movie_title']
```

```python
'''Second we will check the data having Positive profit'''

positive_df = df[df['Profit']>0]
positive_df
```

```python
'''Data Visualization of Budget having positive impact on Profit'''

sns.jointplot(x=positive_df['budget'],y=positive_df['Profit'])
```

```python
'''Top 5 Movies having positive profit'''

df.sort_values(by='imdb_score', ascending=False)['movie_title'][:5]
```

# Task 2 : Find IMDB Top 250.

```python
'''1. Creating column name IMDb_Top_250 and sorting using column imdb_score'''

IMDb_Top_250 = df[df['num_voted_users']>25000].sort_values(by='imdb_score', ascending = False).head(250)
IMDb_Top_250.reset_index(inplace=True,drop=True)

'''2. Adding Column name Rank in IMDb_Top_250'''

IMDb_Top_250['rank'] = IMDb_Top_250['imdb_score'].rank(method='max',ascending=0)


print('Top 250 "Movie_Title" with their "imdb_score" and "rank".\n')

IMDb_Top_250[['movie_title','imdb_score','rank']]
```

# Extract all the movies in the IMDb_Top_250 column which are not in the English language.

```python
'''Top 250 Foreign Lang film'''

Top_Foreign_Lang_Film = IMDb_Top_250[IMDb_Top_250['language'] != 'English']

print('Top 250 Foreign Language film are as follows\n')

Top_Foreign_Lang_Film['movie_title']
```

# Task 3 : Find the best directors.

```python
'''To find top 10 directors'''

'''First we will group dataset by director and find mean of imdb_score '''

top10director = df[['director_name','imdb_score']].groupby('director_name')['imdb_score'].mean().reset_index()

'''Case 1. Sorting the data using imdb_score'''

print('Top 10 directors in case 1 are\n')

top10director.sort_values(by=['imdb_score'],ascending=False).head(10)['director_name']
```

```python
'''Case 2. Sorting the data using imdb_score and director_name'''

print('Top 10 directors in case 2 are\n')

top10director.sort_values(by=['imdb_score','director_name'],ascending=False).head(10)['director_name']
```

# Task 4 : Find popular genres.

```python
'''Grouping by Genres and taking mean of imdb_score and movie_facebook_likes'''

top_10_genres = df.groupby('genres')[['imdb_score','movie_facebook_likes']].mean().reset_index()

'''Sorting data using imdb_score and movie_facebook_likes'''

top_10_genres.sort_values(by=['imdb_score','movie_facebook_likes'],ascending=False,inplace=True)
```

```python
'''Showing Top 10 genres loved by users'''

print('Top 10 genres are as follows \n')

top_10_genres['genres'].head(10)
```

# Task 5 : Identify the actors which have the highest mean.

As we know that variable name num_user_for_reviews has dtype object and these variable is used for finding the top actors hence we will convert the dtype of num_user_for_reviews from object to int.

```python
'''Converting dtype of num_user_for_reviews from object to int'''

df['num_user_for_reviews'] = df['num_user_for_reviews'].apply(lambda x : int(x))

print('The dtype is changed.')
```

```python
'''Grouping the data by actor_1_name and finding the mean of num_user_for_reviews and num_critic_for_reviews'''

actor_highest_mean = df.groupby('actor_1_name')[['num_user_for_reviews','num_critic_for_reviews']].mean().reset_index()

'''Sorting the data using num_user_for_reviews and num_critic_for_reviews'''

actor_highest_mean.sort_values(by=['num_critic_for_reviews','num_user_for_reviews'],ascending=False,inplace=True)

'''Lets try to see top 3 actors'''

print('Overall Top Lead 3 actors are : \n')

actor_highest_mean.head(3)
```

Now as the task is to find who among Meryl Streep, Brad Pitt and Leonardo DiCaprio have highest mean.

```python
combined = actor_highest_mean[((actor_highest_mean['actor_1_name']=='Meryl Streep')|(actor_highest_mean['actor_1_name']=='Brad Pitt')|(actor_highest_mean['actor_1_name']=='Leonardo DiCaprio'))]
combined.reset_index(drop=True, inplace= True)
combined
```

Hence we can see that Leonardo DiCaprio has highest mean.
