Now by studying the data we can say that column 'NAME_CONTRACT_TYPE', 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY' are object dtypes having two values. So we can convert these columns into numeric dtypes by apply and lambda function.

```python
df['NAME_CONTRACT_TYPE'] = df['NAME_CONTRACT_TYPE'].apply(lambda x : 1 if x == 'Cash loans' else 0)
df['CODE_GENDER'] = df['CODE_GENDER'].apply(lambda x : 1 if x == 'M' else 0)
df['FLAG_OWN_CAR'] = df['FLAG_OWN_CAR'].apply(lambda x : 1 if x == 'Y' else 0)
df['FLAG_OWN_REALTY'] = df['FLAG_OWN_REALTY'].apply(lambda x : 1 if x == 'Y' else 0)
```

```python
'''Selecting Categorical column for finding number of Unique Values '''

Categorical_col = df.select_dtypes(include=np.object)
col_name = list(Categorical_col)

print('The Categorical Columns are : ')

col_name
```

```python
'''Counting Number of Unique values present in Categorical Columns'''

for col in col_name:
    print('Column ', col)
    print('Total Number of unique elements ', len(list(Categorical_col[col].unique())))
    print('***************\n')
```

As we can see from above code that each categorical column has more than 5 unique values. Hence we will drop this column.
